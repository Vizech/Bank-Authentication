```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from vecstack import stacking
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
```

```python
link = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
names = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols'
         'Flavanoids', 'Nonflavanoid phenols', 'proanthocyanins',  'Color intensity', 'Hue', '0D280/0D315 of diluted wines', 'proline']

df = pd.read_csv(link, header =None, names=names)
df.head()
```

```python
X_train = X_train.values if isinstance(X_train, (pd.Series, pd.DataFrame)) else X_train
X_test = X_test.values if isinstance(X_test, (pd.Series, pd.DataFrame)) else X_test
```

```python
X_train = X_train.reshape(-1, 1) if X_train.ndim == 1 else X_train
X_test = X_test.reshape(-1, 1) if X_test.ndim == 1 else X_test
```

```python
X = df.iloc[:,1]
y = df[['Class']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Define your models
models = [
    KNeighborsClassifier(n_neighbors=5),
    RandomForestClassifier(n_estimators=100, random_state=0),
    LogisticRegression(max_iter=1000)
]
```

```python
X_train = X_train.values
X_test = X_test.values
```

```python
if X_train.ndim == 1:
    X_train = X_train.reshape(-1, 1)
if X_test.ndim == 1:
    X_test = X_test.reshape(-1, 1)
```

```python
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

```python
# Convert y_train to categorical if it's not already
if isinstance(y_train, (pd.Series, pd.DataFrame)):
    y_train = y_train.values  # Convert to numpy array if needed

# Ensure y_train is of integer type and categorical
y_train = y_train.astype(int)

# Print unique values to verify
print(np.unique(y_train))
```

```python
# Change the mode to 'oof' or 'pred_bag' for regression
S_train, S_test = stacking(
    models,
    X_train,
    y_train,
    X_test,
    regression=True,  # Set to True for regression
    mode='oof_pred_bag',
    needs_proba=False,
    save_dir=None,
    metric=accuracy_score,  # Consider using appropriate regression metric
    n_folds=4,
    stratified=False,  # Stratified is not needed for regression
    shuffle=True,
    random_state=0,
    verbose=2
)
```
