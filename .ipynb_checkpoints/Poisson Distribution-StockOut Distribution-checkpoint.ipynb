```python
import math
from scipy.stats import poisson
```

```python
lambda_ = 2
```

```python
cumulative_prob_0_to_6 = poisson.cdf(6, lambda_)
print("Cumulative Probability P(X <= 6):", cumulative_prob_0_to_6)
```

```python
stockout_prob = 1 - cumulative_prob_0_to_6
print("Stockout Probability P(X > 6):", stockout_prob)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define your data points and parameters
X = [0, 1, 2, 3, 4, 5]
mu = 2
poisson_pd = poisson.pmf(X, mu)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the Poisson PMF
ax.plot(X, poisson_pd, 'o', mfc='r', label='Poisson PMF')  # 'o' for markers; 'mfc' for marker face color

# Add vertical lines
ax.vlines(X, 0, poisson_pd, colors='b', linewidth=1, alpha=0.5)

# Add labels and title
plt.ylabel('Probability', fontsize=18)
plt.xlabel('X = No. of Restaurants', fontsize=18)
plt.title('Poisson Distribution - No. of Restaurants vs Probability', fontsize=18)

# Add legend
ax.legend()

# Show the plot
plt.show()
```
