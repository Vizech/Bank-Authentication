```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.impute import SimpleImputer
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetes = load_diabetes()
```

```python
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
```

```python
df['target'] = diabetes.target
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

for column in df:
    plt.figure()
    df.boxplot([column])
```

```python
Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound: {lower_bound}, upper Bound: {upper_bound}")
```

```python
df_no_outliers = df[(df['bmi'] >= lower_bound) & (df['bmi'] <= upper_bound)]
print("Outliers removed from the 'bmi' column.")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_no_outliers['bmi'])
plt.title("Boxplot of bmi after Outlier Removal")
plt.show()
print(f"New Min: {df_no_outliers['bmi'].min()}, New Max: {df_no_outliers['bmi'].max()}")
```
