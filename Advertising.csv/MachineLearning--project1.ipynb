```python
import pandas as pd
```

```python
df_adv_data = pd.read_csv("Advertising.csv", index_col=0)
```

```python
df_adv_data.head()
```

```python
df_adv_data.size
```

```python
X_features = df_adv_data[['Newspaper','Radio','TV']]
```

```python
Y_target = df_adv_data[['Sales']]
```

```python
Y_target.head()
```

```python
X_features.shape
```

```python
Y_target.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_features,Y_target,random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
y_pred = linreg.predict(X_test)
y_pred
```

```python
from sklearn import metrics
import numpy as np
```

```python
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error (RMSE):", rmse)
```
