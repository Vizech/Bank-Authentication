```python
# Strip any leading/trailing spaces from all column names
df.columns = df.columns.str.strip()
drugs = [f"Drug{x}" for x in range(1, 9)]  # List of drug column names

for drug in drugs:
    if f"{drug}" in df.columns and f"{drug}_sales" in df.columns:
        mean_ratio = np.mean(df[f"{drug}_sales"] / df[drug])
        print(f"Mean of {drug}_sales / {drug} = {mean_ratio}")
    else:
        print(f"Columns {drug} or {drug}_sales not found in DataFrame.")
```

# Weekly Analysis

```python
wdf = df.resample('w').mean()
```

```python
wdf.head()
```

```python
wdf.describe()
```

# Inference from the describe function of weekly analysis
1. Drug4 tops the sales
2. Drug6 sold the least units
3. Drug7 performed the worst in terms of sales even if it is sold close to 6 units.

```python
import pandas as pd

# Define the date range for all weeks you expect to see
all_weeks = pd.date_range(start=wdf.index.min(), end=wdf.index.max(), freq='W-SUN')

# Reindex the DataFrame to include all weeks
wdf_reindexed = wdf.reindex(all_weeks)

# Fill missing values with 0 for Drug1_sales
wdf_filled = wdf_reindexed.fillna(0)

```

```python
import matplotlib.pyplot as plt

# Set figure size and DPI
plt.figure(figsize=(6, 3), dpi=100)

# Rotate the x-axis labels for better readability
plt.tick_params(axis='x', rotation=45)

# Set the title of the plot
plt.title('Drug1 Weekly Sales Overview (All Weeks Included)')

# Plot a bar chart for Drug1 sales
plt.bar(x=wdf_filled.index, height=wdf_filled['Drug1_sales'])

# Show the plot
plt.show()
```

# Inference of Drug1 weekly sales:
1. Best week was from 05/09/2019 & 29/09/2019
2. first week and last week of september made almost the same axis
