```python
unique_countries = list(confirmed_cases['Country/Region'].unique())
unique_countries
```

```python
import operator

country_confirmed_cases = []
no_cases = []

# Calculate the number of confirmed cases for each country
for i in unique_countries:
    cases = latest_confirmed[confirmed_cases['Country/Region'] == i].sum()
    if cases > 0:
        country_confirmed_cases.append(cases)
    else:
        no_cases.append(i)

# Remove countries with no cases
for i in no_cases:
    unique_countries.remove(i)

# Sort countries by confirmed cases in descending order
unique_countries = [k for k, v in sorted(zip(unique_countries, country_confirmed_cases), key=operator.itemgetter(1), reverse=True)]

# Update country_confirmed_cases based on sorted countries
country_confirmed_cases = [latest_confirmed[confirmed_cases['Country/Region'] == unique_countries[i]].sum() for i in range(len(unique_countries))]
```

```python
print('Confirmed Cases by Countries/Regions:')
for i in range(len(unique_countries)):
    print(f'{unique_countries[i]}: {country_confirmed_cases[i]} cases')
```

```python
# Ensure the correct column name is used
unique_provinces = list(confirmed_cases['Province/State'].unique())

# Define outliers
outliers = ['united kingdom', 'denmark', 'france']

# Remove outliers if they exist in unique_provinces
for i in outliers:
    # Convert both to lowercase for case-insensitive comparison, and handle non-string values
    to_remove = next((province for province in unique_provinces if isinstance(province, str) and province.lower() == i), None)
    if to_remove:
        unique_provinces.remove(to_remove)

```

```python
province_confirmed_cases = []
no_cases = []
for i in unique_provinces:
    cases = latest_confirmed[confirmed_cases['Province/State']==i].sum()
    if cases > 0:
        province_confirmed_cases.append(cases)
    else:
        no_cases.append(i)

for i in no_cases:
    unique_provinces.remove(i)
```

```python
for i in range(len(unique_provinces)):
    print(f'{unique_provinces[i]}: {province_confirmed_cases[i]} cases')
```

```python
nan_indices = []

for i in range(len(unique_provinces)):
    if type(unique_provinces[i]) == float:
        nan_indices.append(i)

unique_provinces = list(unique_provinces)
province_confirmed_cases = list(province_confirmed_cases)

for i in nan_indices:
    unique_provinces.pop(i)
    province_confirmed_cases.pop(i)
```
