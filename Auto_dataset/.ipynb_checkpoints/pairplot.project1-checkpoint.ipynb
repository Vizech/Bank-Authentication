```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df_auto_dataset = pd.read_csv('auto.csv')
```

```python
df_auto_dataset.head()
```

```python
def origin(num):
    if num == 1:
        return 'USA'
    elif num == 2:
        return 'Europe'
    else:
        return 'Asia'

df_auto_dataset['origin'] = df_auto_dataset['origin'].apply(origin)
```

```python
df_auto_dataset.head(30)
```

```python
sns.pairplot(df_auto_dataset[['mpg','weight','origin']],hue='origin',size=4)
```

# pie Chart Demo Assignment2

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cause = 'Chronic Disease','Unintentional Injuries','Alzheimers','Influenza and Pneumonia','Sepsis','Others'
percentile = [62,5,4,2,1,26]
```

```python
plt.figure(figsize=(10,10))
explode = (0.05,0,0,0,0,0)
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle=70)
plt.axis('equal')
plt.title('Ohio State - 2012 : leading Cause of Death')
plt.show()
```
