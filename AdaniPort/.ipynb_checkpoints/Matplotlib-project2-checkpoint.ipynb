```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ADANIPORTS.NS.csv',parse_dates = ['Date'])
```

```python
df.head()
```

```python
import seaborn as sns
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()
sns.set_style("darkgrid")
```

```python
# Create a 3D axis
ax = plt.axes(projection='3d')

# Plot the scatter plot
ax.scatter(df.index, df['H-L'], df['100MA'])

# Set axis labels
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')

# Show the plot
plt.show()
```

```python
# Create a 3D axis
ax = plt.axes(projection='3d')

# Plot the scatter plot
ax.scatter(df.index, df['H-L'], df['Volume'])

# Set axis labels
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('Volume')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data
Z1 = np.linspace(0, 10, 100)
X1 = np.cos(Z1)  # Generate cosine values based on Z1
Y1 = np.sin(Z1)  # Generate sine values based on Z1

# Set style
sns.set_style('whitegrid')

# Create a 3D plot
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(X1, Y1, Z1)  # Plot the 3D line

# Show the plot
plt.show()
```

```python
def return_z(x, y):
    return 50 - (x == 2 and y == 2)
```

```python
sns.set_style('whitegrid')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a linear space and generate a meshgrid
x1 = np.linspace(-5, 5, 50)
y1 = np.linspace(-5, 5, 50)
x1, y1 = np.meshgrid(x1, y1)

# Vectorize the return_z function to apply it element-wise to arrays
return_z_vectorized = np.vectorize(return_z)
z1 = return_z_vectorized(x1, y1)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x1, y1, z1, cmap='viridis')

# Show the plot
plt.show()

```
